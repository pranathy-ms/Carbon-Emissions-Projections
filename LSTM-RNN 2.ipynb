```python
import pandas as pd

# Load the CSV file
file_path = 'ghg-emissions-by-sector-stacked.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df.head()
```

```python
df.dropna(inplace=True)
print(len(df['Entity'].unique()))
```

```python
# Define sector columns
sector_columns = [
    'Greenhouse gas emissions from other fuel combustion',
    'Greenhouse gas emissions from bunker fuels',
    'Greenhouse gas emissions from waste',
    'Greenhouse gas emissions from buildings',
    'Greenhouse gas emissions from industry',
    'Fugitive emissions of greenhouse gases from energy production',
    'Greenhouse gas emissions from agriculture',
    'Greenhouse gas emissions from manufacturing and construction',
    'Greenhouse gas emissions from transport',
    'Greenhouse gas emissions from electricity and heat'
]

global_emissions_by_sector = df.groupby('Year')[sector_columns].sum()
global_emissions_by_sector
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Dictionary to store rolling MSE and future predictions for each sector
sector_results = {}

# Process the grouped global emissions data
years = global_emissions_by_sector.index.values  # Extract years as index
future_years = 10  # Number of years to predict into the future

for sector in sector_columns:
    print(f"Processing sector: {sector}")

    # Get the sector data and reshape for scaling
    reg_sector = global_emissions_by_sector[sector].values.reshape(-1, 1)
    
    # Scale the data
    reg_sector_scaled = scaler.fit_transform(reg_sector)

    # Use a window length of 10 for rolling regression
    window_length = 5

    # Prepare the data for LSTM
    X_train = []
    y_train = []
    for i in range(window_length, len(reg_sector_scaled)):
        X_train.append(reg_sector_scaled[i-window_length:i, 0])
        y_train.append(reg_sector_scaled[i, 0])

    X_train, y_train = np.array(X_train), np.array(y_train)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

    # Initialize and compile the LSTM model
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(LSTM(units=50))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Fit the model
    model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=0)

    # Perform rolling prediction
    rolling_predictions = []
    rolling_true_values = []
    rolling_years = years[window_length:]  # Adjusted years for predictions

    for i in range(window_length, len(reg_sector_scaled)):
        X_window = reg_sector_scaled[i-window_length:i].reshape(1, window_length, 1)
        y_window = reg_sector_scaled[i:i+1].reshape(-1, 1)

        # Predict the next value
        y_pred_window = model.predict(X_window)
        rolling_predictions.append(y_pred_window[0, 0])
        rolling_true_values.append(y_window[0, 0])

    # Inverse transform the predictions and true values to original scale
    rolling_predictions_scaled = np.array(rolling_predictions).reshape(-1, 1)
    rolling_predictions_original = scaler.inverse_transform(rolling_predictions_scaled)
    rolling_true_values_scaled = np.array(rolling_true_values).reshape(-1, 1)
    rolling_true_values_original = scaler.inverse_transform(rolling_true_values_scaled)

    # Calculate rolling MSE for the current sector
    rolling_mse = mean_squared_error(rolling_true_values_scaled, rolling_predictions_scaled)


    # Predict the next 10 years
    last_window = reg_sector_scaled[-window_length:].reshape(1, window_length, 1)  # Use the last window to predict
    future_predictions = []
    future_years_list = [years[-1] + i for i in range(1, future_years + 1)]  # Generate future years

    for _ in range(future_years):
        # Predict the next value
        next_pred = model.predict(last_window)
        future_predictions.append(next_pred[0, 0])
        
        # Update the last window with the new prediction
        last_window = np.append(last_window[:, 1:, :], next_pred.reshape(1, 1, 1), axis=1)

    # Inverse transform the future predictions to original scale
    future_predictions_scaled = np.array(future_predictions).reshape(-1, 1)
    future_predictions_original = scaler.inverse_transform(future_predictions_scaled)

    # Store results for the sector
    sector_results[sector] = {
        'rolling_mse': rolling_mse,
        'rolling_predictions': rolling_predictions_original,
        'future_predictions': future_predictions_original,
        'future_years': future_years_list
    }

    # Print rolling MSE
    print(f"Rolling MSE for sector '{sector}': {rolling_mse}")

    # Optional: Plot actual, rolling predictions, and future predictions
    plt.plot(years, reg_sector, label="Actual Data", color="blue")
    plt.plot(rolling_years, rolling_predictions_original, label="Rolling Predictions", color="red")
    plt.plot(future_years_list, future_predictions_original, label="Future Predictions", color="green")
    plt.title(f"Global Greenhouse Gas Emissions Forecast ({sector})")
    plt.xlabel("Year")
    plt.ylabel("Emissions")
    plt.legend()
    plt.show()


```

```python
# Print final rolling MSE and display summary
print("\nSummary of Results:")
for sector, results in sector_results.items():
    print(f"Sector: {sector}")
    print(f"  Rolling MSE: {results['rolling_mse']}")
```

```python
import matplotlib.pyplot as plt
import random

# Initialize the plot
plt.figure(figsize=(25, 15))

# Function to generate a random color
def random_color():
    return (random.random(), random.random(), random.random())

# Loop through each sector and plot the actual data, rolling predictions, and future predictions
for sector in sector_columns:
    print(f"Plotting sector: {sector}")

    # Retrieve the sector data and results
    reg_sector = global_emissions_by_sector[sector].values
    rolling_predictions = sector_results[sector]['rolling_predictions']
    future_predictions = sector_results[sector]['future_predictions']
    future_years_list = sector_results[sector]['future_years']
    
    # Rolling years are adjusted for each sector based on the window length
    rolling_years = years[window_length:]

    # Generate a random color for this sector
    color = random_color()

    # Plot actual data, rolling predictions, and future predictions for the sector
    plt.plot(years, reg_sector, label=f"{sector} - Actual", linestyle='-', color=color)
    plt.plot(rolling_years, rolling_predictions, label=f"{sector} - Rolling Predictions", linestyle='--', color=color)
    plt.plot(future_years_list, future_predictions, label=f"{sector} - Future Predictions", linestyle=':', color=color)

# Add titles and labels
plt.title("Global Greenhouse Gas Emissions Forecast - All Sectors")
plt.xlabel("Year")
plt.ylabel("Emissions")
plt.legend()

# Display the combined plot
plt.show()
```
